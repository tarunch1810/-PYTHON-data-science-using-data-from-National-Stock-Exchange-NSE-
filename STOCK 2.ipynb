```python
import nsepy
import datetime
import pandas as pd
```

```python
def gethdata(stockname):
    today=datetime.date.today()
    duration=45
    start=today+datetime.timedelta(-duration)
    stockdata=nsepy.get_history(symbol=stockname,start=start,end=today)
    return stockdata
df=gethdata('SBIN')
```

```python
df.head()
```

```python
sd=df[['Prev Close','High','Deliverable Volume','VWAP','Low']]
```
